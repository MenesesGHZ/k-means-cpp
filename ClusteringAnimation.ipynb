```python
### GETTING DATA PROPERLY

with open("centroides.txt") as file:
    file = file.read().strip()
    file = [block.strip() for block in file.split("\n")]
    n_steps = len(file)//3
    CENTROIDES = [list(),list(),list()] #centroid_0,centroid_1,centroid_2
    PUNTOS = [list(),list(),list()]  # puntos_0, puntos_1, puntos_2
    for n,centroid in enumerate(file):
        data = centroid.split("|-|")
        try:
            coords = data[0].replace("(","")
            coords = coords.replace(")","")
            coords = coords.split(",")
            coords = [float(coords[0]), float(coords[1])]
            puntos = [int(index) for index in data[1].split(" ")]
        except IndexError:
            print("HAY UN ERROR PERO AUN ASI FUNCIONA")
        finally:
            CENTROIDES[n%3].append(coords)
            PUNTOS[n%3].append(puntos)
            
with open("puntos-500.txt") as file:
    file = file.read().strip()
    file = file.split("\n")
    PUNTOS_COORDS = list()
    for punto in file:
        punto = punto.split(",")
        punto = [int(punto[0]),int(punto[1])]
        PUNTOS_COORDS.append(punto)
    

    
```

```python
### GRAPH
%matplotlib notebook
import matplotlib.pyplot as plt
import time

plt.style.use('dark_background')

size = (500,500) # 500width ,500height
red   = "#FF0000"
green = "#00FF00"
blue  = "#0000FF"

plt.plot([0,500],[0,500])


    
```

```python
for step in range(n_steps):
    plt.cla()

    x_0, y_0 = CENTROIDES[0][step]
    x_1, y_1 = CENTROIDES[1][step]
    x_2, y_2 = CENTROIDES[2][step]
    
    for punto in PUNTOS[0][step]:
        x_p, y_p = PUNTOS_COORDS[punto]
        plt.scatter(x_p, y_p,c=red)
        
    for punto in PUNTOS[1][step]:
        x_p, y_p = PUNTOS_COORDS[punto]
        plt.scatter(x_p, y_p,c=green)
        
    for punto in PUNTOS[2][step]:
        x_p, y_p = PUNTOS_COORDS[punto]
        plt.scatter(x_p, y_p,c=blue)
    
    plt.scatter(x_0, y_0,s=150,c=red,edgecolors='white')
    plt.scatter(x_1, y_1,s=150,c=green,edgecolors='white')
    plt.scatter(x_2, y_2,s=150,c=blue,edgecolors='white')
    
    plt.savefig('samples/sample_{}.png'.format(step))
            
        
```
